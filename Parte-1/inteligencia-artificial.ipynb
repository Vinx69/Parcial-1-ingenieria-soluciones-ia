```python
import os
from openai import OpenAI
from langchain_openai import ChatOpenAI
from langchain_core.chat_history import BaseChatMessageHistory, InMemoryChatMessageHistory
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_core.messages import HumanMessage

import os
import time

print("Librerias importadas!")
```

```python
try:
    llm = ChatOpenAI(
        base_url=os.getenv("OPENAI_BASE_URL"),
        api_key=os.getenv("GITHUB_TOKEN"),
        model="gpt-4.1",
        streaming=True,
        temperature=0.1
    )
    
    print("Modelo configurado")
    print(f"Modelo: {llm.model_name}")
    print(f"Streaming: {llm.streaming}")
    
except Exception as e:
    print(f"✗ Error en configuración: {e}")
    print("Verifica las variables de entorno")
```

```python
prompt = ChatPromptTemplate.from_messages([
    ("system", "Eres un asistente que presta servicios en una empresa de transportes, tu objetivo es saber donde quiere empezar y terminar el pasajero, cuantos pasajeros la fecha y hora del inicio del viaje, cuando te hablen, las primeras preguntas que tienes que hacer son para pedir dicha información. Todo es en la ciudad de Puerto Montt del país de Chile, así que si te dicen un lugar, debe ser de la zona, así que con eso recolectas la información y le vas preguntando al cliente la lozación que creas que pidió para saber que esté correcta y el te dará el veredicto final para aceptar el viaje"),
    MessagesPlaceholder(variable_name="chat_history"),
    ("human", "{input}")
])

chain = prompt | llm

store = {}

def get_session_history(session_id: str):
    if session_id not in store:
        store[session_id] = InMemoryChatMessageHistory()
    return store[session_id]

conversation = RunnableWithMessageHistory(
    chain,
    get_session_history,
    input_messages_key="input",
    history_messages_key="chat_history"
)

def ejemplo_buffer_memory():
    print("-----CONVERSACIÓN CON CLIENTE-----")
    
    session_id = "demo_session"

    try:
        print("--PRIMERA INTERACCIÓN--")
        response1 = conversation.invoke(
            {"input": "hola, mi nombre es Juan y quiero un viaje"},
            config={"configurable": {"session_id": session_id}}
        )
        print("\nhola, mi nombre es Juan y quiero un viaje\n")
        print(f"Respuesta: {response1.content}\n")

        # Segunda interacción
        print("--SEgUNDA INTERACCIÓN--")
        response2 = conversation.invoke(
            {"input": "del terminal hacia el aeropuerto, para el día 08-04-2026, solo somos 2 pasajeros"},
            config={"configurable": {"session_id": session_id}}
        )
        print("\ndel terminal hacia el aeropuerto, para el día 08-04-2026, solo somos 2 pasajeros\n")
        print(f"Respuesta: {response2.content}\n")

        # Tercera interacción
        print("--CUARTA INTERACCIÓN--")
        response3 = conversation.invoke(
            {"input": "disculpa!, sería a las 7:00 de la mañana, gracias"},
            config={"configurable": {"session_id": session_id}}
        )
        print("\ndisculpa!, sería a las 7:00 de la mañana, gracias\n")
        print(f"Respuesta: {response3.content}\n")

        print("--QUINTA INTERACCIÓN--")
        response4 = conversation.invoke(
            {"input": "si, todo correcto, gracias!!"},
            config={"configurable": {"session_id": session_id}}
        )
        print("\nsi, todo correcto, gracias!!\n")
        print(f"Respuesta: {response4.content}\n")


    except Exception as e:
        print(f"Error: {e}")

# Ejecutar
ejemplo_buffer_memory()
```
