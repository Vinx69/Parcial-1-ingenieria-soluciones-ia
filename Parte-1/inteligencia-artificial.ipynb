```python
#imports
import os
from dotenv import load_dotenv
load_dotenv()

from langchain_openai import ChatOpenAI
from langchain_core.chat_history import BaseChatMessageHistory, InMemoryChatMessageHistory
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.runnables.history import RunnableWithMessageHistory
print("Librerias importadas!")
```

```python
#configuración de la API
token = os.environ["GITHUB_TOKEN"]

endpoint = "https://models.github.ai/inference"

model = "openai/gpt-4.1"
```

```python
#configuración del modelo
llm = ChatOpenAI(
    base_url=endpoint,
    api_key=token,
    model=model,
    temperature=0.1,
    streaming=True
)
```

```python
#prompt
prompt = ChatPromptTemplate.from_messages([
    ("system", "Eres un asistente que presta servicios en una empresa de transportes, tu objetivo es saber donde quiere empezar y terminar el pasajero, cuantos pasajeros la fecha y hora del inicio del viaje, cuando te hablen, las primeras preguntas que tienes que hacer son para pedir dicha información. Todo es en la ciudad de Puerto Montt del país de Chile, así que si te dicen un lugar, debe ser de la zona, así que con eso recolectas la información y le vas preguntando al cliente la lozación que creas que pidió para saber que esté correcta y el te dará el veredicto final para aceptar el viaje."),
    MessagesPlaceholder(variable_name="chat_history"),
    ("human", "{input}")
])

chain = prompt | llm
```

```python
#Almacén de historial por sesión

store = {}

def get_session_history(session_id: str) -> BaseChatMessageHistory:
  if session_id not in store:
    store[session_id] = InMemoryChatMessageHistory()
  return store[session_id]
```

```python
#Envolver con memoria

conversation = RunnableWithMessageHistory(
  chain,
  get_session_history,
  input_messages_key="input",
  history_messages_key="chat_history"

)
```

```python
#Función de chat

session_id = "mi_sesion"

def chat(mensaje):
  response = conversation.invoke(
    {"input": mensaje},
    config={"configurable": {"session_id": session_id}}
  )
  print(f"Cliente: {mensaje}")
  print(f"Profesional Vulcanizador: {response.content}")
  print("-" * 40)
```

```python
#chat
chat("hola, mi nombre es Juan y quiero un viaje")

chat("del terminal hacia el aeropuerto, para el día 08-04-2026, solo somos 2 pasajeros")

chat("disculpa!, sería a las 7:00 de la mañana, gracias")

chat("si, todo correcto, gracias!!")
```
