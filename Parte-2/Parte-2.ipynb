```python
#IMPORTS
import os
os.environ["LANGCHAIN_TRACING_V2"] = "false"
os.environ["LANGCHAIN_API_KEY"] = "dummy"

from dotenv import load_dotenv
load_dotenv()

from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage
# Configuración del modelo (igual al chat con ia.)
token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.github.ai/inference"
model_name = "openai/gpt-4.1"

llm = ChatOpenAI(
  base_url=endpoint,
  api_key=token,
  model=model_name,
  temperature=0.1,
  streaming=True
)

print("✓ Modelo listo")
```

```python
#CODIGO
def chat_zero_shot(pregunta):

  prompt = f"""Eres un asistente amable de una empresa de transportes llamada TransportesPardo.
Tarea: Recolectar la información de los viajes de cada cliente, tu objetivo es saber donde quiere empezar y terminar el pasajero, cuantos pasajeros, la fecha y hora del inicio del viaje, cuando te hablen, las primeras preguntas que tienes que hacer son para pedir dicha información. Todo es en la ciudad de Puerto Montt del país de Chile, así que si te dicen un lugar, debe ser de la zona.
Formato de respuesta:
- Datos faltantes: [pregunta para obtener datos faltantes del viaje]
- Validación: [pregunta para validar que la información del cliente sea correcta preguntantdo si la información es correcta]

Consulta del cliente: "{pregunta}"
"""
  response = llm.invoke([HumanMessage(content=prompt)])
  print(f" Cliente   : {pregunta}")
  print(f" Asistente : {response.content}")
  print("-" * 50)



# Prueba
chat_zero_shot("hola, mi nombre es Juan y quiero un viaje del terminal hacia el aeropuerto, para el día 08-04-2026, solo somos 2 pasajeros, sería a las 7:00 de la mañana, gracias")
```

```python
#IMPORTS
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage
import os
import json
import re

# Configurar el modelo
llm = ChatOpenAI(
    base_url=os.getenv("OPENAI_BASE_URL"),
    api_key=os.getenv("GITHUB_TOKEN"),
    model="gpt-4o",
    temperature=0.3  # Menor temperatura para más consistencia
)

print("✓ Modelo configurado para few-shot prompting")
print("✓ Temperature reducida para mayor consistencia")
```

```python
#CODIGO
def chat_few_shot(pregunta):

  prompt = f"""Eres un Asistente virtual amable de una empresa de transportes que se encargará de recibir la información de los viajes, necesitas el lugar de inicio y final, pasajeros y hora, preguntando si la información está correcta para confirmar el viaje.

Responde siguiendo exactamente el estilo de estos ejemplos:

---

Cliente   : hola, mi nombre es Juan y quiero un viaje del terminal hacia el aeropuerto, para el día 08-04-2026, solo somos 2 pasajeros, sería a las 7:00 de la mañana, gracias
 Asistente : - Datos faltantes: ¿Podrías confirmarme si el "terminal" es el Terminal de Buses de Puerto Montt? Además, ¿el aeropuerto es el Aeropuerto El Tepual de Puerto Montt?
- Validación: Entonces, Juan, ¿me confirmas que la información es la siguiente?  
  - Origen: Terminal de Buses de Puerto Montt  
  - Destino: Aeropuerto El Tepual de Puerto Montt  
  - Fecha: 08-04-2026  
  - Hora de inicio: 07:00 AM  
  - Pasajeros: 2  
¿Es correcto?

---

Cliente: hola, mi nombre es Juan y quiero un viaje
Profesional Vulcanizador: ¡Hola Juan! Para poder ayudarte con tu viaje, necesito algunos datos:

1. ¿Desde qué lugar de Puerto Montt quieres iniciar tu viaje?
2. ¿A qué lugar de Puerto Montt te gustaría llegar?
3. ¿Cuántos pasajeros serán?
4. ¿En qué fecha y a qué hora deseas iniciar el viaje?

Por favor, indícame estos detalles para poder continuar con tu solicitud.

---

Cliente: "{pregunta}"

Asistente:"""
  response = llm.invoke([HumanMessage(content=prompt)])
  print(f"Cliente   : {pregunta}")
  print(f"Asistente : {response.content}")
  print("-" * 50)



# Pruebas

chat_few_shot("hola, mi nombre es Juan y quiero un viaje del terminal hacia el aeropuerto, para el día 08-04-2026, solo somos 2 pasajeros, sería a las 7:00 de la mañana, gracias")
```
